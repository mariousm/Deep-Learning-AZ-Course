## Importación de librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## Carga del dataset

```python
df = pd.read_csv("./data/Churn_Modelling.csv")
df.head()
```

```python
X = df.iloc[:, 3:13]
y = df.iloc[:, 13]
print("Variables independientes")
print("**********************************************************************")
print(X.head())
print("\nVariable dependiente")
print("**********************************************************************")
print(y.head())
```

## Preprocesado de datos

```python
# Dummies tanto para Geography y Gender,
# eliminamos una columna de Geography y de Gender para problemas de colinealidad
X = pd.get_dummies(X, drop_first=True)

print("Variables independientes")
print("**********************************************************************")
print(X.head())

# Normalización
sc_X = StandardScaler()
X = sc_X.fit_transform(X)

print("\nVariables independientes normalizadas")
print("**********************************************************************")
print(X[0:5])

# Conjunto de train y de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print("\nConjuntos de train y test")
print("**********************************************************************")
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```
