## Importación de librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix

from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## Carga del dataset

```python
df = pd.read_csv("./data/Churn_Modelling.csv")
df.head()
```

```python
X = df.iloc[:, 3:13]
y = df.iloc[:, 13]
print("Variables independientes")
print("**********************************************************************")
print(X.head())
print("\nVariable dependiente")
print("**********************************************************************")
print(y.head())
```

## Preprocesado de datos

```python
# Dummies tanto para Geography y Gender,
# eliminamos una columna de Geography y de Gender para problemas de colinealidad
X = pd.get_dummies(X, drop_first=True)

print("Variables independientes")
print("**********************************************************************")
print(X.head())

# Normalización
sc_X = StandardScaler()
X = sc_X.fit_transform(X)

print("\nVariables independientes normalizadas")
print("**********************************************************************")
print(X[0:5])

# Conjunto de train y de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print("\nConjuntos de train y test")
print("**********************************************************************")
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Construcción de la RNA

```python
# Inicializamos la red neuronal
rna = Sequential()

# Creamos las capas de la red
rna.add(Dense(units = 6, kernel_initializer = "uniform", activation = "relu", input_dim = 11)) # 1ª capa oculta
rna.add(Dense(units = 6, kernel_initializer = "uniform", activation = "relu")) # 2ª capa oculta
rna.add(Dense(units = 1, kernel_initializer = "uniform", activation = "sigmoid")) # Capa de salida
```

## Entrenamiento de la RNA

```python
# Compilamos la RNA
rna.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])

# Entrenamos
rna.fit(X_train, y_train, batch_size=10, epochs=100)
```

## Predicción de los resultados

```python
# Predicción
y_pred = rna.predict(X_test)
y_pred = (y_pred > 0.5)

# Matriz de confusión
confusion_matrix(y_test, y_pred)
```
